```python
import seaborn as sns
sns.set_theme(style="whitegrid")
df=sns.load_dataset("titanic")
print(df.info())
df.head()
```

```python
df.describe().T
```

# Missing Data Analysis

```python
df["age"].fillna(df.age.mean(),inplace=True)
df["embarked"].fillna(df.embarked.mode()[0],inplace=True)
df["embark_town"].fillna(df["embark_town"].mode()[0],inplace=True)
df.info()
```

# Survival Analysis by Gender and Age:

```python
sns.catplot(data=df,x="alive",hue="who",kind="count")
```

# Passenger Class and Survival

```python
sns.catplot(data=df,x="class",hue="alive",kind="count")
```

# Embarkation Port and Survival

```python
sns.catplot(data=df,x="alive",hue="embark_town",kind="count")
```

# Family Size and Survival

```python
sns.catplot(data=df,x="sibsp",hue="parch",kind="count",col="alive")
```

# Fare and Survival

```python
sns.violinplot(data=df,x="fare",y="alive")
```

# Cabin Location and Survival

```python
sns.countplot(data=df,x="deck",hue="alive")
```

# Survival by Gender and Class

```python
sns.countplot(data=df,x="class",hue="alive")
```

# Age and Gender Interaction

```python
sns.violinplot(data=df,x="age",y="sex")
```

# Children vs. Adults

```python
sns.violinplot(data=df,x="age",y="who",hue="alive")
```
